```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
'''
Este código
apaenas traça um gr[afico
pra teste]
'''
w = np.linspace(-np.pi, np.pi, 100) # base das abicissas dos pontos
h = np.sin(w) # ordenadas calculadas sobre as abscissas
plt.plot(w, h)
plt.grid()
plt.title('Função de transferência')
plt.show()
```

```python
w = np.linspace(0, 10000, 100000) # base das abicissas dos pontos
R = 68e3 # 68 k
L = 10
a = L/R
H = 1/np.sqrt(1 + (a**2)*w**2)
wc = 2000*np.pi


plt.plot(w, H)
wdest = [wc, wc]
hdest = [0, 1/np.sqrt(2) + .05]
plt.plot(wdest, hdest, 'r')
plt.grid()
plt.title('Função de transferência')
plt.show()
```

```python
H_mag_dB = 20 * np.log10(H)

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.semilogx(w, H_mag_dB) # Usando semilogx para escala logarítmica no eixo x
plt.xlabel('Frequência Angular (rad/s)')
plt.ylabel('Magnitude (dB)')
plt.title('Resposta em Magnitude de um Filtro Passa-Baixa RC (em dB)')
plt.grid(True, which="both", ls="--")
plt.axvline(wc, color='red', linestyle='--', label=f'Frequência de Corte (wc = {wc:.2f} rad/s)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do filtro passa-baixa RC
R = 1e3  # Resistência em Ohms
C = 1e-6 # Capacitância em Farads

# Frequência de corte angular (wc = 1 / (R*C))
wc = 1 / (R * C)

# Frequências para o gráfico (em rad/s)
w = np.logspace(-1, 5, 500) # Usando logspace para melhor visualização em escala logarítmica

# Função de transferência do filtro passa-baixa (Magnitude)
H_mag = 1 / np.sqrt(1 + (w/wc)**2)

# Convertendo para dB (20*log10(H_mag))
H_mag_dB = 20 * np.log10(H_mag)

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.semilogx(w, H_mag_dB) # Usando semilogx para escala logarítmica no eixo x
plt.xlabel('Frequência Angular (rad/s)')
plt.ylabel('Magnitude (dB)')
plt.title('Resposta em Magnitude de um Filtro Passa-Baixa RC (em dB)')
plt.grid(True, which="both", ls="--")
plt.axvline(wc, color='red', linestyle='--', label=f'Frequência de Corte (wc = {wc:.2f} rad/s)')
plt.legend()
plt.show()
```
